#Mini-práctica A
**(listas + diccionarios)**: crea un registro de gastos con categorías; al final imprime el total y el total por categoría.


```python
# Crear diccionario de gastos
gastos = {}
```

```python
# Registrar gastos por categoria
salud = [10,20,30,40]
Alimentación = [1,2,3,4]
transporte = [10,10.2,10.3,10.4]
entretenimiento = [100,200,300,400]
```

```python
# Añadir listas al diccionario
gastos["salud"] = salud
gastos["Alimentación"] = Alimentación
gastos["transporte"] = transporte
gastos["entretenimiento"] = entretenimiento
```

```python
# Obtener e imprimir el total por categoría
print("Resumen de gastos por categoría: \n")
for key, value in gastos.items():
  print("El total en",key, "es: $", float(sum(value)))
```

#Mini-práctica C
**(tipos de datos)**: pide 3 números al usuario, guárdalos en lista y calcula promedio; muestra el tipo de cada valor antes y después de convertir.

```python
# Declarar calificaciones

calificaciones = [input("Calificación 1: "),
                  input("Calificación 2: "),
                  input("Calificación 3: ")]

# Revisar el tipo de valor antes de convertir
print(type(calificaciones[0]))

#Convertir y calcular el promedio
promedio = (int(calificaciones[0]) + int(calificaciones[1]) + int(calificaciones[2]))/3

#Imprimir resultado de operación
print("El promedio de tus calificaciones es: "promedio)

# Revisa el tipo de valor después de converti
print(type(promedio))

```
