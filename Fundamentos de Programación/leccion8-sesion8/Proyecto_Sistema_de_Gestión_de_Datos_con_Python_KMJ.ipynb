#### Funciones de Estilo

````python
def estilar_encabezado(encabezado:str):
    """
    Imprime en consola un encabezado estilizado con líneas superior e inferior.
    Args:
        encabezado (str): El texto a mostrar en el encabezado.
    """
    ancho_maximo = 26 # Ancho máximo de la línea
    print("") # Salto de línea
    print("=" * ancho_maximo)
    print(f"{encabezado.center(ancho_maximo)}") # Centra el encabezado en la línea
    print("-"* ancho_maximo)
````

````python
def estilar_pie(pie:str):
    """
    Imprime en consola un pie de sección estilizado con una línea inferior.
    Args:
        pie (str): El texto a mostrar en el pie.
    """
    ancho_maximo = 26 # Ancho máximo del diseño
    print("")
    print(f"{pie.center(ancho_maximo)}") # Centra el contenido en la línea de acuerdo con el ancho establecido
    print("-"*ancho_maximo)
    print("")
````

````python
def centrar_texto(texto:str):
    "Centra un texto"
    print(f"{texto.center(26)}")
````

````python
def estilar_separador():
    """Imprime una línea separadora simple para mejorar la legibilidad."""
    print("-"*26)
    print("")
````

````python
# Traduce el estado de 1, 2 o 3 a "Baja", "Media" o "Alta"
def traducir_prioridad(indice_prioridad:int) -> str:
    """
    Convierte el valor numérico de la prioridad a una cadena de texto legible.
    Args:
        prioridad_int (int): El número de la prioridad (1, 2, o 3).
    Returns:
        str: La representación en texto ('Baja', 'Media', 'Alta') o 'Desconocida'.
    """
    if indice_prioridad == 1:
        return "Baja"
    elif indice_prioridad == 2:
        return "Media"
    elif indice_prioridad == 3:
        return "Alta"
    else:
        return "Prioridad no válida"
````

````python
# Aligera la presentación de la fecha al presentarla en dd-mm-aaaa
def traducir_fecha(fecha:datetime) -> str:
    """
    Formatea un objeto datetime a una cadena de texto en formato 'Día-Mes-Año'.
    Args:
        fecha_datetime (datetime): El objeto de fecha a formatear.
    Returns:
        str: La fecha formateada como una cadena.
    """
    return fecha.strftime("%d-%m-%Y")
````

````python
# Traduce el estado de True o False a "Completado" o "No completado"
def traducir_estado(estado:bool) -> str:
    """
    Convierte el valor booleano del estado de la tarea a una cadena legible.
    Args:
        estado_bool (bool): El estado de la tarea (True para completado).
    Returns:
        str: 'Sí' si es True, 'No' si es False.
    """
    if estado:
        return "Completada"
    else:
        return "No Completada"
````

````python
# Mejora la presentación de las claves estandarizadas de un diccionario
def traducir_característica(caracteristica:str)->str:
    """
    Formatea el nombre de una clave de diccionario para su presentación.
    Reemplaza guiones bajos por espacios y capitaliza la primera letra.
    Args:
        caracteristica (str): El nombre de la clave (ej. 'fecha_de_creacion').
    Returns:
        str: El nombre formateado (ej. 'Fecha de creacion').
    """
    caracteristica_formateada = caracteristica.replace("_"," ").capitalize().strip()
    return caracteristica_formateada
````

````python
# Formatea un encabezado para presentación y muestra las opciones del menu
def mostrar_menu (encabezado:str, opciones:list):
    """
    Muestra un menú genérico y numerado en consola.
    Args:
        titulo (str): El título del menú.
        opciones (list): Una lista de cadenas, donde cada una es una opción.
    """
    estilar_encabezado(encabezado)
    for i, opcion in enumerate(opciones,1):
        print(f"{i}. {opcion}")
````

````python
# Muestra una sola tarea e identifica que tipo de caracteristica es para dar un mejor formato

def mostrar_tarea(tarea:dict | None):
    """
    Muestra en consola los detalles de una única tarea de forma formateada.
    Utiliza funciones de traducción para presentar los valores de forma legible.
    Args:
        tarea (dict): El diccionario de la tarea a mostrar.
    """

    # Mapeo de claves
    formateadores = {
        "prioridad": traducir_prioridad,
        "completado": traducir_estado,
        "fecha_de_creacion": traducir_fecha,
        "fecha_limite": traducir_fecha
    }

    if tarea is None:
        return None

    for caracteristica, contenido in tarea.items():

        formateador = formateadores.get(caracteristica) # Obtiene la función de formateo si existe

        if formateador:
            print(f"{traducir_característica(caracteristica)}: {formateador(contenido)}")
        else:
            print(f"{traducir_característica(caracteristica)}: {contenido}")
````

#### Helpers reutilizables

````python
# Preguntar si es correcto que ponga int o None en esta función auxiliar
def buscar_tarea_id(id:int | None, tareas:list)-> dict | None:
    """
    Busca una tarea en la lista de tareas por su ID.
    Args:
        id_buscado (int): El ID de la tarea a buscar.
        tareas (list): La lista de diccionarios de tareas donde buscar.
    Returns:
        dict | None: El diccionario completo de la tarea si se encuentra,
                     o None si no hay ninguna tarea con ese ID.
    """
    for tarea in tareas:
        if (id == tarea["id"]):
            return tarea
    return None
````

````python
def validador_id(id:int, tareas:list)-> bool:
    """Toma un id y verifica si es un número válido (entero), y si existe dentro de la lista de tareas
    args:
    id
    """
    if buscar_tarea_id(id, tareas) != None:
        return True
    return False
````

````python
def generador_id():
    """
    Generador infinito que produce IDs numéricos secuenciales.
    Cada vez que se itera sobre él (usando next()), devuelve el siguiente
    entero, comenzando desde 1.
    Yields:
        int: El siguiente ID en la secuencia.
    """
    contador_ids = 1
    while True:
        yield contador_ids
        contador_ids += 1

generador_de_id_tareas = generador_id() # Variable para activar el generador de ids, si se llama a generador_id() directamente se llama a la fabrica, no al resultado
````

````python
def entrada_no_vacia(mensaje = "Ingrese el contenido: ")-> str:
    """
    Solicita una entrada al usuario y se asegura de que no esté vacía.
    Permite al usuario cancelar la entrada escribiendo 'cancelar'.
    Args:
        mensaje (str): El mensaje a mostrar al usuario para solicitar la entrada.
    Returns:
        str | None: La cadena de texto ingresada por el usuario (sin espacios extra),
                    o None si el usuario escribe 'cancelar'.
    """
    while True:
        entrada_raw = input(mensaje)

        if entrada_raw == None:
            continue
        try:
            entrada_limpia = entrada_raw.strip().lower()
            return entrada_limpia
        except ValueError:
            print("Debe contener al menos un fragmento de texto")
````

````python
def seleccionar_opcion_menu(titulo:str, opciones:list)-> int | None:
    """
    Muestra un menú y gestiona la selección del usuario.
    Asegura que la opción sea un número válido dentro del rango o que se cancele.
    Args:
        titulo (str): El título a pasar a mostrar_menu().
        opciones (list): La lista de opciones a pasar a mostrar_menu().
    Returns:
        int | None: El número (1-based) de la opción elegida, o None si se canceló.
    """
    mostrar_menu(titulo, opciones)
    try:
        eleccion_raw = input(f"Selecciona una opción (1-{len(opciones)}): ")

        if eleccion_raw.lower() == "cancelar":
            return None

        eleccion_int = int(eleccion_raw)

        if 1 <= eleccion_int <= len(opciones):
            return eleccion_int
        else:
            print(f"Error: Opción inválida. Por favor, ingresa un número entre 1 y {len(opciones)}.")
    except ValueError:

        print("Error: Entrada inválida. Por favor, ingresa un número.")
    except Exception as e:
        print(f"Ocurrió un error inesperado: {e}")
````

##### Obtener entradas del usuario validadas

````python
def obtener_id_del_usuario()->int | None:
    """
    Solicita al usuario un ID numérico y lo valida.
    El bucle continúa hasta que se ingresa un número entero válido o se cancela.
    Args:
        mensaje (str): El mensaje a mostrar al usuario.
    Returns:
        int | None: El ID como un entero si es válido, o None si el usuario cancela.
    """
    while True:
        entrada_procesada = entrada_no_vacia("Ingresa el número de id de la nota (o escribe 'cancelar'): ")

        if entrada_procesada == "cancelar":
            return None

        try:
            entrada_limpia = int(entrada_procesada)
            return entrada_limpia
        except ValueError:
            print("El id solo contiene numeros enteros sin símbolos")
````

````python
def obtener_descripcion_del_usuario()->str | None:
    descripcion = entrada_no_vacia("Ingrese la descripción de la nota (o escribe 'cancelar'):")

    if descripcion == "cancelar":
        return None

    return descripcion
````

````python
def obtener_fecha_limite_del_usuario()-> datetime | None:
    """
    Para simplificar la creación le daremos al usuario la opción de ingresar o no alguna fecha y en caso de que se equivoque al momento de formatear, usaremos la fecha por default.
    """
    print("El siguiente campo es opcional, sino lo ingresa con el formato indicado se le asignará una fecha de dos días a partir de la creación de la nota")

    fecha_limite = datetime.now() + timedelta(days=2)
    entrada_fecha_limite = input("Ingresa la fecha límite (YYYY-MM-DD o escribe 'cancelar'): ")# **timedelta** permite representar una diferencia de tiempo

    if entrada_fecha_limite == "cancelar":
        return None

    if entrada_fecha_limite:
        # **Try y except** se usan para manejar errores.
        try:
            fecha_limite = datetime.strptime(entrada_fecha_limite, "%Y-%m-%d") # Lo que hace **strptime()** es "parsear" o interpretar esa cadena de texto y la convierte en un objeto de tipo datetime
        except ValueError:
            print("Formato de fecha inválido. Se usará la fecha predeterminada de dos días.")
    return fecha_limite
````

````python
def obtener_prioridad_del_usuario()->int | None:

    while True:
        entrada_raw = entrada_no_vacia("Ingrese el número que representa la prioridad (1 = Baja, 2 = Media, 3 = Alta, o 'cancelar'): ")

        if entrada_raw == "cancelar":
            return None
        try:
            prioridad_int = int(entrada_raw)
            if prioridad_int > 0 and prioridad_int <= 3:
                return prioridad_int
            else:
                estilar_pie("Intente nuevamente, se esperaba un número ya sea 1 o 2 o 3")

        except ValueError:
            print("Se esperaba un número entero: 1, 2 o 3")
````

````python
def obtener_entrada_o_cancelar(input_func):
    """
    Función auxiliar que llama a una función de entrada de usuario y gestiona la cancelación (devolviendo None).
    """
    valor = input_func()
    if valor is None:
        estilar_pie("Creación de tarea cancelada por el usuario.")
    return valor
````

````python
def obtener_tarea(tareas:list)->dict | None:

    tarea_id = obtener_entrada_o_cancelar(obtener_id_del_usuario)

    if tarea_id is None:
        return None

    tarea_encontrada = buscar_tarea_id(tarea_id, tareas)

    if tarea_encontrada is None:
        estilar_pie(f"No se encontró ninguna tarea con el ID '{tarea_id}'.")
        return None

    return tarea_encontrada
````

````python
def obtener_estado_del_usuario()-> bool | None:
    while True:
        entrada_raw = entrada_no_vacia("Ingrese el estado de la Tarea: 1 para Completado y 0 para No completado (o 'cancelar'): ")
        if entrada_raw == "cancelar":
            return None
        if entrada_raw == "1":
            return True
        elif entrada_raw == "0":
            return False
        else:
            estilar_pie("Opcion No Válida, intente nuevamente\n")
````

##### Gestionar acciones del usuario

````python
def gestionar_crear_tarea():
    """
    Función de alto nivel que gestiona la creación y adición de una nueva tarea.
    Llama a crear_tarea() y, si tiene éxito, añade la nueva tarea a la lista.
    Args:
        tareas (list): La lista principal de tareas a la que se añadirá la nueva.
    """
    tarea_a_añadir = crear_tarea()
    if tarea_a_añadir is not None:
        tareas.append(tarea_a_añadir)
        estilar_pie("Tarea añadida con éxito")
    else:
        pass
````

````python
def gestionar_eliminar_tarea():
    tarea_id = obtener_entrada_o_cancelar(obtener_id_del_usuario)

    if tarea_id is None:
        return

    tarea_encontrada = buscar_tarea_id(tarea_id, tareas)

    if tarea_encontrada is None:
        estilar_pie(f"No se encontró ninguna tarea con el ID '{tarea_id}'.")
        return

    eliminado_con_exito = eliminar_tarea(tarea_encontrada, tareas)

    if eliminado_con_exito:
        estilar_pie("Proceso de eliminación finalizado")
    else:
        estilar_pie("El proceso de eliminación no resultó en la eliminación de la tarea.")
````

### **Crear Tarea**
Define una función para crear una nota con los campos indicados; la fecha de creación e id y estado: false, deberá realizarse en automático.

* Descripción
* Fecha límite (default = dos días después de la fecha de creación)(opcional)
* Prioridad (default = 2)(opcional)
* Prioridad:
  * 1. Baja
  * 2. Media
  * 3. Alta

````python
def crear_tarea()-> dict | None:
    """
    Gestiona el flujo para crear una nueva tarea pidiendo los datos al usuario.
    Permite la cancelación en cualquier paso de la entrada de datos.
    Returns:
        dict | None: Un nuevo diccionario de tarea con todos los datos si se completó,
                     o None si el usuario canceló el proceso.
    """
    estilar_encabezado("Crear Tarea")

    # Para revisar si la entrada del usuario no fue cancelada
    descripcion = obtener_entrada_o_cancelar(obtener_descripcion_del_usuario)
    if descripcion is None : return None

    fecha_limite = obtener_entrada_o_cancelar(obtener_fecha_limite_del_usuario)
    if fecha_limite is None : return None

    prioridad = obtener_entrada_o_cancelar(obtener_prioridad_del_usuario)
    if prioridad is None : return None

    tarea = {
        "id":next(generador_de_id_tareas),
        "descripcion":descripcion,
        "completado": False,
        "fecha_de_creacion":datetime.now(),
        "fecha_limite":fecha_limite,
        "prioridad": prioridad
        }

    estilar_pie("Tarea creada con éxito")

    return tarea
````

###**Consultar Tarea**

````python
def consultar_tarea (tareas:list)-> dict | None:
    """
    Gestiona el flujo completo para consultar una tarea.
    Pide un ID al usuario, busca la tarea y, si la encuentra, la muestra.
    Args:
        tareas (list): La lista de tareas donde se realizará la búsqueda.
    Returns:
        dict | None: La tarea encontrada si el proceso fue exitoso, o None
                     si la tarea no se encontró o el usuario canceló.
    """
    estilar_encabezado("Consultar Tarea")
    id_raw = obtener_entrada_o_cancelar(obtener_id_del_usuario)
    if id_raw is None:
        return None
    tarea = buscar_tarea_id(id_raw, tareas)

    if tarea is None:
        estilar_pie("No hay tareas con ese id")
        return None
    mostrar_tarea(tarea)
    return tarea
````

#### *Editar Tarea*

````python
# Editar Tarea
def editar_tarea(tarea_a_editar: dict):
    """
    Muestra un menú para editar los campos de una tarea específica.
    Permite al usuario editar múltiples campos de la misma tarea en un bucle
    hasta que decida volver al menú principal.
    Args:
        tarea_a_editar (dict): El diccionario de la tarea a modificar.
    """
    editado_con_exito = False # Para saber si se hizo algún cambio

    while True: # Bucle para permitir editar varios campos seguidos
        estilar_encabezado(f"Editando Tarea ID: {tarea_a_editar['id']}")
        estilar_pie("Tarea actual:")
        mostrar_tarea(tarea_a_editar) # Muestra el estado actual de la tarea

        opciones_edicion = [
            "Modificar Descripción",
            "Cambiar Estado",
            "Actualizar Fecha Límite",
            "Ajustar Prioridad",
            "Volver al Menú Principal"
        ]

        eleccion_int = seleccionar_opcion_menu("¿Qué campo deseas editar?", opciones_edicion)

        if eleccion_int is None or eleccion_int == 5:
            if editado_con_exito:
                estilar_pie("Cambios guardados. Regresando al menú principal.")
            else:
                estilar_pie("No se realizaron cambios. Regresando al menú principal.")
            break

        # Lógica para cada opción de edición
        if eleccion_int == 1: # Editar Descripción
            nueva_descripcion = obtener_descripcion_del_usuario()
            if nueva_descripcion is not None:
                tarea_a_editar["descripcion"] = nueva_descripcion
                editado_con_exito = True

        elif eleccion_int == 2: # Editar Estado
            nuevo_estado = obtener_estado_del_usuario()
            if nuevo_estado is not None:
                tarea_a_editar["completado"] = nuevo_estado
                editado_con_exito = True

        elif eleccion_int == 3: # Editar Fecha Límite
            nueva_fecha_limite = obtener_fecha_limite_del_usuario()
            if nueva_fecha_limite is not None:
                tarea_a_editar["fecha_limite"] = nueva_fecha_limite
                editado_con_exito = True

        elif eleccion_int == 4: # Editar Prioridad
            nueva_prioridad = obtener_prioridad_del_usuario()
            if nueva_prioridad is not None:
                tarea_a_editar["prioridad"] = nueva_prioridad
                editado_con_exito = True
````

#### *Modificar Descripción*

````python
def modificar_descripcion(tarea:dict)-> bool:
    estilar_encabezado("Modificar descripción")
    nueva_descripcion = obtener_descripcion_del_usuario()
    if nueva_descripcion:
        tarea["descripcion"] = nueva_descripcion
        return True
    return False
````

#### *Cambiar Estado*

````python
def cambiar_estado(tarea:dict)-> bool:
    estilar_encabezado("Cambiar Estado")
    nuevo_estado = obtener_estado_del_usuario()
    if nuevo_estado is not None:
        tarea["completado"] = obtener_estado_del_usuario()
        return True
    return False
````

#### *Actualizar Fecha Límite*

````python
def actualizar_fecha_limite(tarea:dict)->bool:
    estilar_encabezado("Actualizar Fecha límite")
    nueva_fecha_limite = obtener_fecha_limite_del_usuario()
    if nueva_fecha_limite:
        tarea['fecha_limite'] = nueva_fecha_limite
        return True
    return False
````

#### *Ajustar Prioridad*

````python
def ajustar_prioridad(tarea:dict| None)-> bool:
    estilar_encabezado("Ajustar Prioridad")

    if tarea == None:
        return False
    nueva_prioridad = obtener_prioridad_del_usuario()
    if nueva_prioridad is not None:
        tarea["prioridad"] = nueva_prioridad
        return True
    return False
````

#### *Mostrar Menú Actualizar*

````python
def inicia_menu_actualizar(tareas:list):
    """
    Inicia el flujo para actualizar una tarea.
    Primero, busca la tarea que el usuario desea editar y, si se encuentra,
    pasa el control a la función editar_tarea().
    Args:
        tareas (list): La lista principal de tareas.
    """
    while True:
        estilar_encabezado("Actualizar una Tarea")
        tarea_a_editar = consultar_tarea(tareas)

        if tarea_a_editar is None:
            estilar_pie("Regresando al menu principal")
            return

        editar_tarea(tarea_a_editar)
````

###**Eliminar Tarea**

````python
def eliminar_tarea(tarea_a_eliminar:dict, tareas:list)-> bool:
    """
    Gestiona la eliminación de una tarea específica, pidiendo confirmación.
    Args:
        tarea_a_eliminar (dict): La tarea específica a eliminar.
        tareas (list): La lista principal de tareas de donde se eliminará.
    Returns:
        bool: True si la tarea fue eliminada con éxito, False en caso contrario.
    """
    estilar_encabezado("Eliminar Tarea")

    # Muestra la tarea que se va a eliminar
    mostrar_tarea(tarea_a_eliminar)

    # Confirmación del usuario
    confirmacion_raw = entrada_no_vacia("¿Estás seguro de que deseas eliminar esta tarea? (s/n o 'cancelar'): ")

    if confirmacion_raw == "cancelar":
        estilar_pie("Eliminación de tarea cancelada por el usuario.")
        return False
    elif confirmacion_raw.lower() != "s": # Si no es 's' (sí)
        estilar_pie("Eliminación de tarea no confirmada. Operación cancelada.")
        return False

    try:
        tareas.remove(tarea_a_eliminar)
        estilar_pie(f"La tarea con el id:'{tarea_a_eliminar['id']}' eliminada con éxito.")
        return True
    except ValueError:
        estilar_pie(f"Error: La tarea con el id:'{tarea_a_eliminar['id']}' no se encontró en la lista de tareas. No se pudo eliminar.")
        return False
    except Exception as e:
        estilar_pie(f"Ocurrió un error inesperado al intentar eliminar la tarea: {e}")
        return False

````

## **Función Integradora**

````python
def iniciar_tasky():
    """
    Función principal que ejecuta el bucle del menú principal del programa.
    """
    while True:
        opcion = mostrar_menu_principal()

        if opcion == 1:
            ver_tareas()
        elif opcion == 2:
             gestionar_crear_tarea()
        elif opcion == 3:
            consultar_tarea(tareas)
        elif opcion == 4:
            inicia_menu_actualizar(tareas)
        elif opcion == 5:
            gestionar_eliminar_tarea()
        elif opcion == 6:
            estilar_pie("Gracias por usar Tasky")
            break
        else:
            print("Opción no válida, intente nuevamente")

iniciar_tasky()
````
