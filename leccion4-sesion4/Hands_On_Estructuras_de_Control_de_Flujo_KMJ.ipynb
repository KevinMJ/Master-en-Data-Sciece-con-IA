## **SELECCIÓN**

Una **estructura de selección** es una **declaración condicional**, el cual permite que diferentes secuencias de código se ejecuten dependiendo de si una **condición booleana** especificada se evalúa como **verdadera o falsa.**

La sintaxis básica de una condicional es la siguiente:

```
if condicion1:
    # código a ejecutar si la condicion1 es verdadera
elif condicion2: (opcional)
    # código a ejecutar si la condicion1 es falsa pero la condicion2 es verdadera
else:
    # código a ejecutar si tanto la condicion1 como la condicion2 son falsas
```

### **EJEMPLO 1:**

Determinar calificación de un examen:

````python
# Definir variable
calificacion = 78
````

````python
# Definir condicional
if calificacion >= 90:
  print("Excelente")
elif calificacion >= 80:
  print("Bueno")
elif calificacion >= 70:
  print("Regular")
else:
  print("Malo")
````

## **ITERACIÓN**

En Python, una **estructura de iteración** es un **ciclo** que generalmente se refiere a dos tipos de **estructuras de control de flujo**: los bucles **for** y **while**.


*   **Bucle For**: El bucle for se usa para **recorrer sobre una secuencia** (que puede ser una lista, una tupla, un diccionario, un conjunto o una cadena) **o cualquier otro objeto iterable**. Cada elemento de la secuencia se procesa en el cuerpo del bucle, uno tras otro.

*   **Bucle While**: El bucle while se usa para **ejecutar repetidamente un bloque de código** siempre que una condición especificada sea verdadera.

### **CICLO FOR**

También conocido como **bucle for**, es un tipo de estructura de control de flujo en Python que permite repetir un bloque de código un número determinado de veces.

El ciclo for tiene la siguiente estructura:

```
for variable in secuencia:
    # cuerpo del ciclo

```

#### **EJEMPLO 1:**

Sumar los números del 1 a 10:

````python
# realiza el ciclo for
suma = 0
for numero in range(1,11):
  suma += numero
print(suma)
````

#### **EJEMPLO 2:**

Insertar numeros secuenciales dentro de una lista:

````python
# definir variables
lista = []
inicio = int(input("Ingrese el número de inicio del rango: "))
fin = int(input("Ingrese el número de final del rango: "))
````

````python
# realizar ciclo for
for numero in range(inicio, fin + 1):
  lista.append(numero)
print(lista)
````

#### **EJEMPLO 3:**

Imprimir las claves y valores de un diccionario:

````python
# define tu variable
estudiantes = {
    "nombre": "carlos",
    "edad": 22,
    "carrera": "Ingenieria Civil"
}
````

````python
# realiza el ciclo for
for clave, valor in estudiantes.items():
  print(clave + ": " + str(valor))
````

### **CICLO WHILE**

También conocido como **bucle while**, es un tipo de estructura de control de flujo en Python que permite repetir un bloque de código mientras se cumpla una condición específica.

El ciclo for tiene la siguiente estructura:

```
while condicion:
    # código a ejecutar mientras se cumpla la condición
```

#### **EJEMPLO 1:**

Validar una cadena:

````python
# Definir variables
usuarios = {"Pedro": "1234", "Maria": "abcd", "Paula": "A1b2c3e4"}
login = True

# realiza el ciclo while
while login:
  usuario = input("Ingrese su usuario: ")
  contrasena = input("Ingrese su contraseña: ")

  if usuario in usuarios and usuarios[usuario] == contrasena:
    print("Bienvenido " + usuario)
    login = False
  else:
    print("Usuario o contraseña incorrectos. Intente nuevamente.")
````

#### **EJEMPLO 2:**

Recorrer las opciones de un menú:

````python
# Definir variables
opciones = ["Ver Perfil", "Editar Perfil", "Cerrar Sesión"]
indice = 0
````

````python
# Realizar ciclo while

while (indice < len(opciones)):
  print(f"{indice + 1}. {opciones[indice]}")
  indice += 1
````

#### **EJEMPLO 3:**

Calcular la suma de los valores de un diccionario:

````python
# define tus variables
ventas = {
    "enero": 1500,
    "febrero": 4000,
    "marzo": 3500
}

claves = list(ventas.keys())
indice = 0
total = 0
````

````python
# realiza el ciclo while
while indice < len(claves):
  total += ventas[claves[indice]]
  indice += 1
print("El total de ventas de todos los meses fue de: ")
print(total)
````

**1. Crea la Agenda:** Construye la agenda con los registros utilizados en la práctica anterior.

````python
# Definir agenda
agenda = [{'nombre': 'dek', 'número_de_telefono': '12', 'correo': 'dek@wer.com'}, {'nombre': 'dekdos', 'número_de_telefono': '1234', 'correo': 'dekdos@wer.com'}, {'nombre': 'dektres', 'número_de_telefono': '12345', 'correo': 'dektres@wer.com'}]
````

**2. Imprime todos los contactos de la agenda:**

   * Realiza un ciclo for que recorra y muestre cada contacto de la agenda.
   * Dentro del ciclo, muestra el número de contacto por cada iteración.

````python
# Mostrar todos los registros de la agenda

# variable para contar el número de contactos
contador_contactos = 1

  # Iterar sobre cada persona en la agenda
for persona in agenda:

  # imprimir el siguiente formato "==== contacto # ===="
  print(f"========== contacto {contador_contactos} ==========")

  # Iterar sobre cada par de clave valor de cada persona
  for key, value in persona.items():
    print(f"{key} : {value}")

  # Actualizar el valor del contador
  contador_contactos += 1
````

**3. Crea un contacto de la agenda:**

   * Pide al usuario que ingrese el nombre, teléfono y correo para un contacto.
   * Guarda estos datos en un diccionario llamado contacto.
   * Imprime el mensaje "Contacto creado con éxito".

````python
# Crear contacto
contacto = {"nombre" :input("¿cuál es tu nombre?"),
              "número_de_telefono" :input("¿cuál es tu número de teléfono?"),
              "correo" :input("¿cuál es tu dirección de correo?")
              }

agenda.append(contacto)
print ("Contacto creado con éxito")
````

**4. Consulta un contacto de la agenda:**

   * Pide al usuario que ingrese el nombre del contacto a consultar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces imprime el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".

````python
# Consultar contacto
contacto_a_buscar = input("Ingresa el nombre del contacto a consultar ")
encontrado = False
for contacto in agenda:
  if contacto["nombre"] == contacto_a_buscar:
    print("Contacto encontrado:")
    print(f"Nombre: {contacto['nombre']}")
    print(f"Teléfono: {contacto['número_de_telefono']}")
    print(f"Correo: {contacto['correo']}")
    encontrado = True
    break

if not encontrado:
    print("Contacto no encontrado")

````

**5. Actualiza un contacto de la agenda:**

   * Pide al usuario que ingrese el nombre del contacto a actualizar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces imprime el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".
   * Crea un menú a través de un ciclo while que muestre las siguientes opciones:
   
   ```
   Menú actualizar contacto:
   
   1. Actualizar nombre
   2. Actualizar teléfono
   3. Actualziar correo
   4. Salir del menú
   ```
   * Solicita al usuario una opción mostrada en el menú, si el usuario no coloca la opción correcta, entonces muestra el mensaje "Contacto no encontrado".
   * Dada la opción escojida para modificar, ingresa ya sea el nuevo nombre, el nuevo teléfono o el nuevo correo del contacto encontrado.
   * Por cada registro nuevo ingresado, imprime de nueva forma el contacto para visualizar los cambios.
   * Define en la última opción un `break` para salir del menú.

````python
# Actualizar contacto
contacto_a_actualizar_nombre = input("Contacto a actualizar ")
contacto_encontrado_obj = None # Variable para almacenar el contacto si se encuentra

# Imprime la información del contacto si es que la encuentra
for contacto in agenda:
  if contacto["nombre"] == contacto_a_actualizar_nombre:
    contacto_encontrado_obj = contacto # Asignamos el contacto encontrado a esta variable
    print("=====================")
    print("Contacto encontrado:")
    print(f"Nombre: {contacto_encontrado_obj['nombre']}")
    print(f"Teléfono: {contacto_encontrado_obj['número_de_telefono']}")
    print(f"Correo: {contacto_encontrado_obj['correo']}")
    break

# Decide si acceder al menu dependiendo si encontramos al contacto_a_actualizar
if contacto_encontrado_obj is None:
  # Si no encontramos al contacto, se imprime lo siguiente y no hay continuación
  print("Contacto no encontrado")
else:
  # Si encontramos al contacto entonces ofrecemos el siguiente menú de opciones
  while True:
    print("""
    Menú actualizar contacto:

    1. Actualizar nombre
    2. Actualizar teléfono
    3. Actualizar correo
    4. Salir del menú
    """)

    opcion = input("Selecciona una opcion en el menu ")

    if opcion == "1":
      nuevo_nombre = input("Ingresa el nuevo nombre ")
      contacto_encontrado_obj["nombre"] = nuevo_nombre
      print("=====================")
      print("Contacto actualizado:")
      print(f"Nombre: {contacto_encontrado_obj['nombre']}")
      print(f"Teléfono: {contacto_encontrado_obj['número_de_telefono']}")
      print(f"Correo: {contacto_encontrado_obj['correo']}")

    elif opcion =="2":
      nuevo_telefono = input("Ingresa el nuevo teléfono ")
      contacto_encontrado_obj["número_de_telefono"] = nuevo_telefono
      print("=====================")
      print("Contacto actualizado:")
      print(f"Nombre: {contacto_encontrado_obj['nombre']}")
      print(f"Teléfono: {contacto_encontrado_obj['número_de_telefono']}")
      print(f"Correo: {contacto_encontrado_obj['correo']}")

    elif opcion =="3":
      nuevo_correo = input("Ingresa el nuevo correo ")
      contacto_encontrado_obj["correo"] = nuevo_correo
      print("=====================")
      print("Contacto actualizado:")
      print(f"Nombre: {contacto_encontrado_obj['nombre']}")
      print(f"Teléfono: {contacto_encontrado_obj['número_de_telefono']}")
      print(f"Correo: {contacto_encontrado_obj['correo']}")

    elif opcion =="4":
      print("Saliendo del menú de actualización")
      break
    else:
      print("=====================")
      print("Opción no válida. Por favor, selecciona una opción válida.")



````

**6. Elimina un contacto de la agenda:**

   * Pide al usuario que ingrese el nombre del contacto a eliminar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces elimina el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".

````python
# Eliminar contacto
contacto_a_eliminar_nombre = input("Contacto a eliminar: ")
eliminado = False

i = 0
while i < len(agenda):
  if agenda[i]["nombre"] == contacto_a_eliminar_nombre:
    print("=====================")
    print("Contacto encontrado:")
    print(f"Nombre: {agenda[i]['nombre']}")
    print(f"Teléfono: {agenda[i]['número_de_telefono']}")
    print(f"Correo: {agenda[i]['correo']}")

    del agenda[i] #Elimina el contacto por indice
    eliminado = True
    print("\n=====================")
    print(f"Contacto '{contacto_a_eliminar_nombre}' eliminado con éxito.")
    break

  i+= 1

if not eliminado:
  print("=====================")
  print("Contacto no encontrado")

# Para ver la agenda después del proceso
print("\nAgenda después del proceso:\n")
contador_contactos = 1
for contacto in agenda:
  print(f"========== contacto {contador_contactos} ==========")
  for key, value in contacto.items():
    print(f"{key} : {value}")
  contador_contactos += 1
  print("")
````
