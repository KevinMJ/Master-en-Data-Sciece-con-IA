````python
# Esta función es para distribuir el toral de una comida en restaurantes

def calcular_total(orden, tasa_imp, porcentaje_propina):
  subtotal = sum(orden.values())
  impuestos = subtotal * tasa_imp
  total_con_impuestos = subtotal + impuestos
  propina = total_con_impuestos * porcentaje_propina
  total_final = total_con_impuestos + propina

  return total_final
````

````python
def leer_grandes_registros(archivo_simulado, limite= None):
  """Simula la lectura eficiente de un gran archivo,
   en lugar de cargar todo en memoria devuelve un registro a la vez"""

  datos =[f"registro {i:04d}"  for i in range(1, 100001)]
  for i, registro, in enumerate(datos):
    if limite and i >=limite:
      return
    yield registro
````

### ¿Qué hace yield?
Cuando ves yield en una función, esa función se convierte automáticamente en una función generadora. Las funciones generadoras son diferentes de las funciones normales:

1. **No devuelven un valor y terminan:** En lugar de return y salir, yield "pausa" la ejecución de la función y devuelve un valor.
2. **Recuerdan su estado:** Cuando la función generadora es llamada de nuevo, no empieza desde el principio. ¡Retoma la ejecución justo donde la dejó después del último yield! Mantiene en memoria el estado de todas sus variables locales
3. **Generan valores bajo demanda (lazy evaluation):** Los valores se producen uno a uno, solo cuando se necesitan. Esto es muy eficiente para secuencias grandes o infinitas, porque no se construye toda la lista en memoria de golpe.

Piensa en yield como una return especial que, en lugar de terminar la función, la "congela", devuelve un valor y luego puede ser "descongelada" para continuar desde el mismo punto.

#### Ventajas de `yield`
* Eficiencia de memoria
* Rendimiento
* Secuencias infinitas

> En pocas palabras, yield te permite escribir funciones que se comportan como iteradores, produciendo una secuencia de valores uno por uno, en lugar de calcularlos todos de golpe.

````python
import time


def medir_tiempo(func):
  "Este decorador mide el tiempo de ejecucion"
  def wrapper(*args, **kwargs):
    inicio = time.time()
    resultado = func(*args, **kwargs)
    fin = time.time()
    print(f"{func.__name__} tardó {fin - inicio: .4f}s")
    return resultado
  return wrapper
````

````python

@medir_tiempo
def procesar_data(n):
  "simular el procesamiento"
  return sum(i**2 for i in range(n))

resultado = procesar_data(100)
print(f"resultado: {resultado}")
````

````python
def registrar_ventas(inv: dict, producto: str, unidades: int=1) -> str:
  "Descontar unidades vendidas de inventario y retorna un mensaje"
  if producto not in inv:
    return f"error: '{producto}' no existe en el inventario"
  disponible = inv[producto]["cantidad"]
  if unidades > disponible:
    return f"Stock insuficiente. Se piden {unidades} de {producto} pero solo hay {disponible} disponibles."
  inv[producto]["cantidad"] -= unidades
  return f"Ventas Ok: {unidades}x{producto} (quedan {inv[producto]['cantidad']})"
````

````python
def recorrer_lista(opciones: list):
    """
    Imprime una lista de opciones numeradas.
    Args:
        opciones (list): La lista de cadenas de texto a imprimir como opciones.
    """
    for i, opcion in enumerate(opciones, 1):
        print(f"{i}. {opcion}")
````

````python
def recorrer_diccionario(datos: dict):
    """
    Imprime de forma legible los pares clave-valor de un diccionario.
    Args:
        datos (dict): El diccionario cuyos contenidos se van a imprimir.
    """
    for clave, valor in datos.items():
        # Mejorar la clave: reemplazar guiones bajos, capitalizar
        clave_formateada = clave.replace("_", " ").capitalize()
        print(f"{clave_formateada}: {valor}")
````

````python
def validar_entrada_no_vacia(entrada, mensaje_error="La entrada no puede estar vacia. Inténtalo de nuevo")-> bool:
    """
    Valida que una cadena de texto no esté vacía y no contenga solo espacios.
    Retorna True si es válida, False en caso contrario.
    """
    if not entrada or entrada.strip() == "":
        print(f"Error: {mensaje_error}")
        return False
    return True
````

````python
def obtener_nombre_valido(mensaje="Ingrese el nombre: "):
    """
    Solicita y valida un nombre.
    Debe contener solo letras y no estar vacío.
    Retorna el nombre válido como str.
    """
    while True:
        entrada_raw = input(mensaje) # Leemos la entrada cruda
        nombre = confirm_action(entrada_raw)

        if nombre is None:
            return None # Propagamos la cancelación

        if not validar_entrada_no_vacia(nombre,"El campo nombre no puede estar vacío"):
            continue
        if all(c.isalpha() or c.isspace() for c in nombre):
            return nombre
        else:
            print("El nombre debe contener solo letras y espacios")
````

````python
def obtener_telefono_valido(mensaje="Ingrese un número de teléfono: "):
    """
    Solicita y valida un número de teléfono.
    Debe ser un número entero y no estar vacío.
    Retorna el número de teléfono válido como int.
    """
    while True:
        entrada_raw = input(mensaje)
        telefono_str = confirm_action(entrada_raw) # Recibe None, "", o el valor limpio

        if telefono_str is None:
            return None # Propaga la cancelación

        if not validar_entrada_no_vacia(telefono_str,"El número de teléfono no puede estar vacio"):
            continue
        try:
            numero_telefono = int(telefono_str)
            return numero_telefono
        except ValueError:
            print("El teléfono debe ser un número válido, inténtalo de nuevo")
````

````python
def obtener_correo_valido(mensaje="Ingrese el correo: "):
    """
    Solicita y valida un correo.
    Debe contener "@" y no estar vacío.
    Retorna el correo válido como str.
    """

    while True:
        entrada_raw = input(mensaje)
        correo = confirm_action(entrada_raw)

        if correo is None:
            return None # Propaga la cancelación

        if not validar_entrada_no_vacia(correo, "El campo correo no debe estar vacio"):
            continue
        if "@" in correo:
            return correo
        else:
            print('El formato del correo no es válido, no olvides incluir el "@" ')
````

````python
def confirm_action(entrada_raw:str):
    """
    Verifica si la entrada del usuario indica una cancelación.
    Retorna None si la entrada es "cancelar" (ignorando mayúsculas/minúsculas y espacios),
    de lo contrario retorna la entrada limpia.
    """
    entrada_limpia = entrada_raw.strip()

    if entrada_limpia.lower() == "cancelar":
        return None # Señal de cancelación
    return entrada_limpia.strip()
````

**5. Define una función para consultar un contacto de la agenda:**

````python
# Crear función consultar contacto
def consultar_contacto_por_nombre(nombre_buscar, agenda):
    """
    Busca un contacto en la agenda por su nombre.
    Si lo encuentra, imprime sus detalles y retorna el diccionario del contacto.
    Si no lo encuentra, imprime un mensaje y retorna None.
    Args:
        nombre_buscar (str): El nombre del contacto a buscar.
        agenda (list): La lista de diccionarios que representa la agenda de contactos.
    Returns:
        dict or None: El diccionario del contacto si es encontrado, de lo contrario None.
    """

    # Normalizar el nombre la búsqueda se vuelve más amigable al permitir uso de mayúsculas y minúsculas
    nombre_estadarizado = nombre_buscar.strip.lower()

    for contacto in agenda:
        if contacto.get("nombre","").lower() == nombre_estadarizado:
            # Contacto encontrado
            estilar_encabezado("Contacto Encontrado")
            recorrer_diccionario(contacto)
            return contacto
    print(f"\nContacto '{nombre_buscar}' no encontrado en la agenda.")
    estilar_separador()
    return None

````

**7. Define una función para eliminar un contacto de la agenda:**

   * Crea una función que pida al usuario que ingrese el nombre del contacto a eliminar.
   * Realiza un ciclo for para la búsqueda del nombre dentro de la agenda.
   * Sí el ciclo encuentra la búsqueda, entonces elimina el contacto encontrado.
   * De lo contrario, entonces imprime "Contacto no encontrado".

````python
def eliminar_contacto(contacto_a_eliminar: dict) -> bool:
    """
    Elimina un contacto específico de la agenda, previa confirmación del usuario.
    Esta función recibe un diccionario que representa el contacto a eliminar.
    Muestra los detalles del contacto y solicita al usuario una confirmación.
    Si el usuario confirma, el contacto es removido de la agenda.
    Maneja la cancelación por parte del usuario y posibles errores durante la eliminación.
    Args:
        contacto_a_eliminar (dict): El diccionario del contacto que se desea eliminar.
    Returns:
        bool: True si el contacto fue eliminado con éxito.
              False si el usuario canceló la operación o si ocurrió un error.
    """
    estilar_encabezado("Eliminar Contacto")

    # Muestra el contacto que se va a eliminar
    print("Se va a eliminar el siguiente contacto:")
    recorrer_diccionario(contacto_a_eliminar)

    # Confirmación del usuario
    confirmacion_raw = input("\n¿Estás seguro de que deseas eliminar este contacto? (s/n o 'cancelar'): ")
    confirmacion = confirm_action(confirmacion_raw)

    if confirmacion is None: # Si el usuario escribió 'cancelar'
        estilar_pie("Eliminación de contacto cancelada por el usuario.")
        return False
    elif confirmacion.lower() != "s": # Si no es 's' (sí)
        estilar_pie("Eliminación de contacto no confirmada. Operación cancelada.")
        return False

    try:
        agenda.remove(contacto_a_eliminar)
        estilar_pie(f"Contacto '{contacto_a_eliminar['nombre']}' eliminado con éxito.")
        return True
    except ValueError:
        estilar_pie(f"Error: El contacto '{contacto_a_eliminar['nombre']}' no se encontró en la agenda. No se pudo eliminar.")
        return False
    except Exception as e:
        estilar_pie(f"Ocurrió un error inesperado al intentar eliminar el contacto: {e}")
        return False
````

**8. Junta todas las funciones del menú CRUD dentro de una función:**

   * Crea una función crear un menú a través de un ciclo while.
   * Solicita al usuario una opción mostrada en el menú, si el usuario no coloca la opción correcta, entonces muestra el mensaje "Opción no encontrada".
   * Dada la opción escojida, ingresa a una función definida con anterioridad que corresponda a su operación a realizar.

````python
# Crear función menú agenda
def menu_agenda():
    """
    Gestiona la interacción con el menú principal de la agenda.
    """
    estilar_encabezado("Agendinsky")
    while True:

        opcion_elegida = mostrar_menu_principal()

        if opcion_elegida == 1:
            mostrar_agenda()
        elif opcion_elegida ==2:
            crear_contacto()
        elif opcion_elegida == 3:
            nombre = input("Ingresa el nombre del contacto a consultar: ")
            consultar_contacto_por_nombre(nombre,agenda)
        elif opcion_elegida == 4:
            actualizar_contacto()
        elif opcion_elegida == 5:
            nombre = input("Ingresa el nombre del contacto a consultar: ")
            contacto_encontrado = consultar_contacto_por_nombre(nombre)
            eliminar_contacto(contacto_encontrado)
        elif opcion_elegida == 6:
            estilar_pie("Saliendo de las opciones del menú CRUD - Agenda")
            break
        else:
            print("Opcion no valida, vuelva a intentar ")
````
